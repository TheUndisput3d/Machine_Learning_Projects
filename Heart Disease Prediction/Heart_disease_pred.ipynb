```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
#loading the csv file to Pandas DataFrame
Heart_dis = pd.read_csv("/content/heart_disease_data.csv")
```

```python
#splitting data into features and target
X = Heart_dis.drop(columns="target", axis=1)
Y = Heart_dis["target"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, Y.shape, X_train.shape , X_test.shape, Y_train.shape , Y_test.shape)
```

#Data standardization

```python
scaler = StandardScaler()
std_X_train = scaler.fit_transform(X_train)
std_X_test = scaler.transform(X_test)
```

#Training the model


logistic regression


```python
model = LogisticRegression()

#trianing the LogisticREgression model with training data
model.fit(X_train, Y_train)
```

#Model Evaluation


```python
#accuracy on training data
y_pred_train = model.predict(X_train)
print("predictions:",y_pred_train)
acc_scr = accuracy_score(y_pred_train, Y_train)
print("The accuracy of the model on the training data is :", acc_scr)
```

```python
#accuracy on testing data
y_pred = model.predict(X_test)
print("predictions:",y_pred)
acc_scr = accuracy_score(y_pred, Y_test)
print("The accuracy of the model on the testing data is :", acc_scr)
```
