```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinson_data = pd.read_csv("/content/parkinsons.data")
```

```python
X = parkinson_data.drop(columns=["name", 'status'], axis=1)
Y = parkinson_data["status"]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3 )
print(X.shape, X_train.shape, X_test.shape, Y.shape, Y_train.shape, Y_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

```python
model = svm.SVC(kernel="linear")
model.fit(X_train, Y_train)
```

```python
#prediction on training data
train_data_pred = model.predict(X_train)
print(train_data_pred)
acc_scr_train = accuracy_score(Y_train, train_data_pred )
print(acc_scr_train)
```

```python
#prediction on testing data
test_data_pred = model.predict(X_test)
print(test_data_pred)
acc_scr_test = accuracy_score(Y_test, test_data_pred )
print(acc_scr_test)
```
